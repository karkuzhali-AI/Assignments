1. What is the difference between a function and a method in Python?
In Python, functions and methods are both callable entities, but they differ in their context and usage. Here’s a breakdown of the key differences:
1. Definition:
Function: A function is defined using the def keyword and can exist independently, not necessarily associated with a particular object or class.
def my_function():
    print("This is a function.")
Method: A method is a function that is associated with an object or a class. In Python, methods are defined inside classes.
Class MyClass:
    def my_method(self):
        print("This is a method."
2. Calling Context:
Function: A function can be called on its own, without requiring any object.
my_function()
Method: A method is called on an instance of a class (or the class itself, in the case of class methods).
obj = MyClass()
obj.my_method()  # Method is called on the object instance.
3.Implicit Argument:
Function: Functions do not take any special implicit arguments (unless explicitly defined).
Method: Instance methods automatically take the instance of the class (commonly self) as their first argument. This allows methods to access and modify the object's attributes.
Example:
class MyClass:
    def my_method(self):
        print("Method called on instance:", self)

obj = MyClass()
obj.my_method()  # `self` is passed implicitly
4.Types:
Function: A regular Python function is of type function.
Method: A method is an instance of the class it is associated with. Python recognizes it as either bound method (if called on an object) or unbound method (in older Python versions, if accessed via the class).
5.Binding:
Function: A function is not bound to any object.
Method: A method is bound to an instance (in the case of instance methods) or the class itself (in the case of class or static methods).
Types of Methods:
Instance Method: A method that operates on the object (instance) of the class.
Class Method: A method bound to the class rather than the instance, using @classmethod and taking cls as its first argument.
Static Method: A method that doesn’t operate on the class or its instances, marked with @staticmethod and doesn’t take self or cls.
Summary:
A function is an independent piece of code that performs a task and can be used anywhere.
A method is a function that is part of a class and typically works with the class's instances or the class itself.

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
In Python, map(), reduce(), and filter() are built-in functions that work with iterables. They help perform transformations and reductions efficiently, but each serves a distinct purpose:
1.Map()
numbers = [1, 2, 3, 4]
result = map(lambda x: x * 2, numbers)  # Doubles each number
print(list(result))  # Output: [2, 4, 6, 8]
2.reduce()
from functools import reduce
numbers = [1, 2, 3, 4]
result = reduce(lambda x, y: x + y, numbers)  # Sums all the numbers
print(result)  # Output: 10
3.filter()
numbers = [1, 2, 3, 4]
result = filter(lambda x: x % 2 == 0, numbers)  # Filters out odd numbers
print(list(result))  # Output: [2, 4]
4.
Key Differences:
map() transforms all elements of an iterable by applying a function.
reduce() aggregates all elements into a single result.
filter() extracts elements that meet a condition, as specified by a function returning True or False.
